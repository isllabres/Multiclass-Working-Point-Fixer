## Tutorial: Multiclass Working Point Fixer with Random Data

```python
from multiclass_wp_fixer.classification.multiclass_working_point_fixer import multiclass_Working_Point_Fixer
from multiclass_wp_fixer.example_data.load_example_data import generate_random_multiclass_data
```

```python
y_train_true, y_train_pred, y_test_true, y_test_pred = generate_random_multiclass_data(
    num_samples=1000,
    num_classes=10,
    val_size=0.2)
```

```python

# Define model parameters
model_params = {
    "input_size": y_train_pred.shape[1],
    "learning_rate": 1e-1,
    "momentum": 0.0,
    "centered": False,
    "num_epochs": 50,
    "multiclass_f1_average": 'macro',
    "train_batch_size": 128,
    "val_batch_size": 128
}

# Initialize the working point fixer
wp_fixer = multiclass_Working_Point_Fixer(model_params=model_params)

# Fit the model
wp_fixer.fit(
    x_tr=y_train_pred,
    y_tr=y_train_true,
    x_tst=y_test_pred,
    y_tst=y_test_true,
    verbose=False,
    plot=True)
```
